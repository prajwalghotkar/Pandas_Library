# Create Series

```python
import pandas as pd
series = pd.Series()
print(series)
```

```python
import numpy as np
arr1 = np.array([10,12,14,16,17,19,21])
series = pd.Series(arr1)
print(series)
```

```python
arr1 = np.array([10,20,30,40,50,60,70,80])
series = pd.Series(arr1)
print(series)
# Extracting the specific element
print(series[::-1])
```

```python
print(series[1:5])
```

```python
print(series[1:8:2])
```

```python
# I wnat to series from dictionary

dict1 = {'Q1':10,'Q2':20,'Q3':30,'Q4':40}
print(dict1)
series = pd.Series(dict1)
print(series)
```

```python
print(series['Q2':'Q4'])
```

```python
print(series['Q3':'Q4'])
```

```python
print(series['Q1':'Q2'])
```

# Update Values in the Series

```python
dict1 = {'Q1':1,'Q2':2,'Q3':3,'Q4':4}
series=pd.Series(dict1)
print(series)
series[1] = 20
print('\n')
print(series)
```

```python
dict1 = {'Q1':10,'Q2':20,'Q3':30,'Q4':50}
series = pd.Series(dict1)
print(series)
print('\n')
series[3] = 300
print(series)
```

# Data Frames

```python
# How to Series convert into DataFrame

import numpy as np
import pandas as pd
list1 = [10,20,30,40,50]
series = pd.Series(list1)
print(series)
print('\n')
print('####### Data Frame ########')
dataframe = pd.DataFrame(series)
display(dataframe)
```

```python
import pandas as pd

data1 = {
    'Name': pd.Series(['Prajwal', 'Jully', 'Lilly', 'Virsika'], index=['a', 'b', 'c', 'd']),
    'Age': pd.Series([21, 23, 24, 21], index=['a', 'b', 'c', 'd']),
    'Salary': pd.Series(['$98798789', '$4354543', '$234432', '$24767823','$8797979'], index=['a', 'b', 'c', 'd','e'])
}

dataframe1 = pd.DataFrame(data1)

# Add new column
dataframe1['Location'] = pd.Series(['USA', 'Brazil', 'UK', 'London'], index=['a', 'b', 'c', 'd'])

# Display the DataFrame
display(dataframe1)
```

```python
import pandas as pd

data1 = {
    'Name': pd.Series(['Prajwal', 'Jully', 'Lilly', 'Virsika'], index=['a', 'b', 'c', 'd']),
    'Age': pd.Series([21, 23, 24, 21], index=['a', 'b', 'c', 'd']),
    'Salary': pd.Series(['$98798789', '$4354543', '$234432', '$24767823'], index=['a', 'b', 'c', 'd'])
}

dataframe1 = pd.DataFrame(data1)

# Add new column
dataframe1['Location'] = pd.Series(['USA', 'Brazil', 'UK', 'London'] , index=['a', 'b', 'c', 'd'])

new_data = pd.Series(['Itally','Brazil'],index=['e','f'])

dataframe1['Location'] = pd.concat([dataframe1['Location'],new_data])
display(dataframe1)
```
